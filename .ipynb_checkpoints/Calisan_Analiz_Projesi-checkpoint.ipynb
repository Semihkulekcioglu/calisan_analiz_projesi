## 1Ô∏è‚É£ K√ºt√ºphaneleri ƒ∞√ße Aktarma


```python
# Veri manip√ºlasyonu i√ßin
import pandas as pd
import numpy as np

# G√∂rselle≈ütirme i√ßin
import matplotlib.pyplot as plt
import seaborn as sns

# ƒ∞statistiksel analizler i√ßin
from scipy import stats

# Uyarƒ±larƒ± gizle
import warnings
warnings.filterwarnings('ignore')

# Grafik ayarlarƒ±
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
%matplotlib inline

# Pandas g√∂r√ºnt√ºleme ayarlarƒ±
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print("‚úÖ T√ºm k√ºt√ºphaneler ba≈üarƒ±yla y√ºklendi!")
```

## 2Ô∏è‚É£ Veri Setini Olu≈üturma

Ger√ßek√ßi bir ≈üirket √ßalƒ±≈üan veri seti olu≈üturacaƒüƒ±z.


```python
# Random seed - tekrar √ºretilebilir sonu√ßlar i√ßin
np.random.seed(42)

# Veri seti boyutu
n_samples = 500

# √áalƒ±≈üan verileri
data = {
    'Calisan_ID': range(1001, 1001 + n_samples),
    'Isim': [f'√áalƒ±≈üan_{i}' for i in range(1, n_samples + 1)],
    'Departman': np.random.choice(['IT', 'Satƒ±≈ü', 'Muhasebe', 'ƒ∞K', 'Pazarlama'], n_samples),
    'Yas': np.random.randint(22, 60, n_samples),
    'Deneyim_Yil': np.random.randint(0, 25, n_samples),
    'Maas': np.random.randint(15000, 85000, n_samples),
    'Performans_Puani': np.random.uniform(1, 10, n_samples).round(2),
    'Egitim_Saati': np.random.randint(0, 120, n_samples),
    'Proje_Sayisi': np.random.randint(0, 15, n_samples),
    'Izin_Gun': np.random.randint(0, 30, n_samples)
}

# DataFrame olu≈ütur
df = pd.DataFrame(data)

# Maa≈üƒ± daha ger√ßek√ßi yapmak i√ßin deneyim ve performansa g√∂re ayarlama
df['Maas'] = (df['Maas'] + 
              df['Deneyim_Yil'] * 1500 + 
              df['Performans_Puani'] * 2000).astype(int)

# Bazƒ± eksik deƒüerler ekleyelim (ger√ßek hayat senaryosu)
df.loc[np.random.choice(df.index, 10), 'Performans_Puani'] = np.nan
df.loc[np.random.choice(df.index, 15), 'Egitim_Saati'] = np.nan

print(f"‚úÖ Veri seti olu≈üturuldu: {df.shape[0]} satƒ±r, {df.shape[1]} s√ºtun")
print(f"\nüìä ƒ∞lk 5 kayƒ±t:")
df.head()
```

## 3Ô∏è‚É£ Veri Seti Hakkƒ±nda Genel Bilgiler


```python
# Veri seti bilgileri
print("üìã VERƒ∞ SETƒ∞ Bƒ∞LGƒ∞LERƒ∞")
print("="*60)
df.info()
```

```python
# ƒ∞statistiksel √∂zet
print("\nüìä ƒ∞STATƒ∞STƒ∞KSEL √ñZET")
print("="*60)
df.describe()
```

```python
# Eksik deƒüer kontrol√º
print("\n‚ùì EKSƒ∞K DEƒûER ANALƒ∞Zƒ∞")
print("="*60)
eksik_degerler = df.isnull().sum()
eksik_oran = (df.isnull().sum() / len(df) * 100).round(2)

eksik_df = pd.DataFrame({
    'Eksik Sayƒ±': eksik_degerler,
    'Eksik Oran (%)': eksik_oran
})

print(eksik_df[eksik_df['Eksik Sayƒ±'] > 0])
```

## 4Ô∏è‚É£ Veri Temizleme ve √ñn ƒ∞≈üleme


```python
# Eksik deƒüerleri doldurma
print("üîß Eksik deƒüerler dolduruluyor...\n")

# Performans puanƒ±nƒ± medyan ile doldur
df['Performans_Puani'].fillna(df['Performans_Puani'].median(), inplace=True)

# Eƒüitim saatini ortalama ile doldur
df['Egitim_Saati'].fillna(df['Egitim_Saati'].mean(), inplace=True)

print("‚úÖ Eksik deƒüerler dolduruldu!")
print(f"\nYeni eksik deƒüer sayƒ±sƒ±: {df.isnull().sum().sum()}")
```

```python
# Yeni √∂zellikler olu≈üturma (Feature Engineering)
print("\nüéØ Yeni √∂zellikler olu≈üturuluyor...\n")

# Ya≈ü gruplarƒ±
df['Yas_Grubu'] = pd.cut(df['Yas'], 
                          bins=[0, 30, 40, 50, 100], 
                          labels=['Gen√ß', 'Orta', 'Kƒ±demli', '√úst D√ºzey'])

# Performans kategorisi
df['Performans_Kategori'] = pd.cut(df['Performans_Puani'],
                                    bins=[0, 4, 7, 10],
                                    labels=['D√º≈ü√ºk', 'Orta', 'Y√ºksek'])

# Deneyim seviyesi
df['Deneyim_Seviye'] = pd.cut(df['Deneyim_Yil'],
                               bins=[-1, 3, 8, 15, 30],
                               labels=['Junior', 'Mid', 'Senior', 'Expert'])

print("‚úÖ Yeni √∂zellikler eklendi!")
print(f"\nG√ºncel s√ºtunlar: {list(df.columns)}")
```

## 5Ô∏è‚É£ Temel ƒ∞statistiksel Analizler

### 5.1 Merkezi Eƒüilim √ñl√ß√ºleri


```python
print("üìä MAA≈û ƒ∞STATƒ∞STƒ∞KLERƒ∞")
print("="*60)
print(f"Ortalama Maa≈ü: {df['Maas'].mean():,.2f} TL")
print(f"Medyan Maa≈ü: {df['Maas'].median():,.2f} TL")
print(f"Mod (En Sƒ±k G√∂r√ºlen): {df['Maas'].mode()[0]:,.2f} TL")
print(f"Standart Sapma: {df['Maas'].std():,.2f} TL")
print(f"Minimum Maa≈ü: {df['Maas'].min():,.2f} TL")
print(f"Maximum Maa≈ü: {df['Maas'].max():,.2f} TL")
print(f"Maa≈ü Aralƒ±ƒüƒ±: {df['Maas'].max() - df['Maas'].min():,.2f} TL")
print(f"\n√áarpƒ±klƒ±k (Skewness): {df['Maas'].skew():.2f}")
print(f"Basƒ±klƒ±k (Kurtosis): {df['Maas'].kurtosis():.2f}")
```

```python
print("\nüìä PERFORMANS ƒ∞STATƒ∞STƒ∞KLERƒ∞")
print("="*60)
print(f"Ortalama Performans: {df['Performans_Puani'].mean():.2f}")
print(f"Medyan Performans: {df['Performans_Puani'].median():.2f}")
print(f"Standart Sapma: {df['Performans_Puani'].std():.2f}")
print(f"\nPerformans Daƒüƒ±lƒ±mƒ±:")
print(df['Performans_Kategori'].value_counts())
```

### 5.2 Departman Bazlƒ± Analizler


```python
print("üè¢ DEPARTMAN BAZLI ƒ∞STATƒ∞STƒ∞KLER")
print("="*60)

# Departman bazlƒ± grupla
departman_analiz = df.groupby('Departman').agg({
    'Maas': ['mean', 'median', 'std', 'min', 'max'],
    'Performans_Puani': ['mean', 'std'],
    'Deneyim_Yil': 'mean',
    'Calisan_ID': 'count'
}).round(2)

departman_analiz.columns = ['Ort_Maa≈ü', 'Med_Maa≈ü', 'Std_Maa≈ü', 'Min_Maa≈ü', 'Max_Maa≈ü',
                            'Ort_Performans', 'Std_Performans', 'Ort_Deneyim', '√áalƒ±≈üan_Sayƒ±']

departman_analiz
```

```python
# En y√ºksek ortalama maa≈üa sahip departman
en_yuksek_maas_dept = departman_analiz['Ort_Maa≈ü'].idxmax()
print(f"\nüí∞ En y√ºksek ortalama maa≈ü: {en_yuksek_maas_dept} departmanƒ±")
print(f"   Ortalama maa≈ü: {departman_analiz.loc[en_yuksek_maas_dept, 'Ort_Maa≈ü']:,.2f} TL")

# En y√ºksek performansa sahip departman
en_yuksek_perf_dept = departman_analiz['Ort_Performans'].idxmax()
print(f"\n‚≠ê En y√ºksek ortalama performans: {en_yuksek_perf_dept} departmanƒ±")
print(f"   Ortalama performans: {departman_analiz.loc[en_yuksek_perf_dept, 'Ort_Performans']:.2f}")
```

### 5.3 Korelasyon Analizi


```python
# Sayƒ±sal s√ºtunlar i√ßin korelasyon matrisi
numeric_cols = ['Yas', 'Deneyim_Yil', 'Maas', 'Performans_Puani', 
                'Egitim_Saati', 'Proje_Sayisi', 'Izin_Gun']

correlation_matrix = df[numeric_cols].corr()

print("üìà KORELASYON ANALƒ∞Zƒ∞")
print("="*60)
print("\nMaa≈ü ile diƒüer deƒüi≈ükenler arasƒ±ndaki korelasyon:")
print(correlation_matrix['Maas'].sort_values(ascending=False))
```

## 7Ô∏è‚É£ Outlier (Aykƒ±rƒ± Deƒüer) Tespiti ve Analizi


```python
def detect_outliers_iqr(data, column):
    """
    IQR (Interquartile Range) y√∂ntemi ile outlier tespiti
    """
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    
    return outliers, lower_bound, upper_bound

# Maa≈ü i√ßin outlier analizi
maas_outliers, maas_lower, maas_upper = detect_outliers_iqr(df, 'Maas')

print("üîç MAAS AYKIRI DEƒûER ANALƒ∞Zƒ∞ (IQR Y√∂ntemi)")
print("="*60)
print(f"Alt Sƒ±nƒ±r: {maas_lower:,.2f} TL")
print(f"√úst Sƒ±nƒ±r: {maas_upper:,.2f} TL")
print(f"\nToplam Aykƒ±rƒ± Deƒüer Sayƒ±sƒ±: {len(maas_outliers)}")
print(f"Aykƒ±rƒ± Deƒüer Oranƒ±: {len(maas_outliers) / len(df) * 100:.2f}%")

if len(maas_outliers) > 0:
    print("\nüìã Aykƒ±rƒ± Deƒüerlere Sahip ƒ∞lk 10 √áalƒ±≈üan:")
    print(maas_outliers[['Isim', 'Departman', 'Maas', 'Deneyim_Yil', 'Performans_Puani']].head(10))
```

## 8Ô∏è‚É£ Pandas ƒ∞leri Seviye ƒ∞≈ülemler


```python
# Pivot Table - Departman ve Deneyim Seviyesine G√∂re Ortalama Maa≈ü
pivot_maas = df.pivot_table(values='Maas', 
                            index='Departman', 
                            columns='Deneyim_Seviye', 
                            aggfunc='mean')

print("üìä PIVOT TABLE: Departman x Deneyim Seviyesi - Ortalama Maa≈ü")
print("="*80)
print(pivot_maas.round(0))

# G√∂rselle≈ütir
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_maas, annot=True, fmt='.0f', cmap='YlOrRd', cbar_kws={'label': 'Maa≈ü (TL)'})
plt.title('Departman ve Deneyim Seviyesine G√∂re Ortalama Maa≈ü', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# GroupBy ile √ßoklu aggregation
print("\nüìä DEPARTMAN BAZLI DETAYLI ANALƒ∞Z")
print("="*80)

detayli_analiz = df.groupby('Departman').agg({
    'Calisan_ID': 'count',
    'Maas': ['mean', 'median', 'std', 'min', 'max'],
    'Performans_Puani': ['mean', 'std'],
    'Deneyim_Yil': ['mean', 'min', 'max'],
    'Proje_Sayisi': 'sum',
    'Egitim_Saati': 'sum'
}).round(2)

detayli_analiz.columns = ['_'.join(col).strip() for col in detayli_analiz.columns.values]
detayli_analiz
```

```python
# En iyi performans g√∂steren √ßalƒ±≈üanlarƒ± bulma
print("\n‚≠ê TOP 10 EN Y√úKSEK PERFORMANSLI √áALI≈ûAN")
print("="*80)
top_performers = df.nlargest(10, 'Performans_Puani')[['Isim', 'Departman', 'Performans_Puani', 'Maas', 'Deneyim_Yil']]
top_performers
```

```python
# Filtreleme √∂rnekleri
print("\nüéØ Fƒ∞LTRELEME √ñRNEKLERƒ∞")
print("="*80)

# 1. Y√ºksek performanslƒ± ve y√ºksek maa≈ülƒ± √ßalƒ±≈üanlar
yuksek_kalite = df[(df['Performans_Puani'] >= 8) & (df['Maas'] >= df['Maas'].quantile(0.75))]
print(f"\n1Ô∏è‚É£ Y√ºksek performans + Y√ºksek maa≈ü: {len(yuksek_kalite)} √ßalƒ±≈üan")

# 2. IT departmanƒ±ndaki senior √ßalƒ±≈üanlar
it_senior = df[(df['Departman'] == 'IT') & (df['Deneyim_Seviye'] == 'Senior')]
print(f"2Ô∏è‚É£ IT Senior √ßalƒ±≈üanlar: {len(it_senior)} ki≈üi")
print(f"   Ortalama maa≈ü: {it_senior['Maas'].mean():,.2f} TL")

# 3. D√º≈ü√ºk performanslƒ± ama y√ºksek maa≈ülƒ± √ßalƒ±≈üanlar (dikkat gerektiren)
anomali = df[(df['Performans_Puani'] < 5) & (df['Maas'] > df['Maas'].mean())]
print(f"\n3Ô∏è‚É£ ‚ö†Ô∏è Anomali durumu (d√º≈ü√ºk perf + y√ºksek maa≈ü): {len(anomali)} √ßalƒ±≈üan")
if len(anomali) > 0:
    print(anomali[['Isim', 'Departman', 'Performans_Puani', 'Maas']].head())
```

## 9Ô∏è‚É£ Numpy ile Matematiksel ƒ∞≈ülemler


```python
print("üî¢ NUMPY ƒ∞LE ƒ∞STATƒ∞STƒ∞KSEL HESAPLAMALAR")
print("="*80)

# Array'e √ßevir
maas_array = df['Maas'].to_numpy()
performans_array = df['Performans_Puani'].to_numpy()

# Temel istatistikler
print(f"\nMaa≈ü Array ≈ûekli: {maas_array.shape}")
print(f"Maa≈ü Ortalamasƒ± (np.mean): {np.mean(maas_array):,.2f} TL")
print(f"Maa≈ü Medyanƒ± (np.median): {np.median(maas_array):,.2f} TL")
print(f"Maa≈ü Std Sapmasƒ± (np.std): {np.std(maas_array):,.2f} TL")
print(f"Maa≈ü Varyansƒ± (np.var): {np.var(maas_array):,.2f}")

# Y√ºzdelikler
print(f"\nüìä Maa≈ü Y√ºzdelikleri:")
percentiles = [10, 25, 50, 75, 90]
for p in percentiles:
    value = np.percentile(maas_array, p)
    print(f"   {p}. Y√ºzdelik: {value:,.2f} TL")

# Korelasyon (numpy ile)
correlation = np.corrcoef(maas_array, performans_array)[0, 1]
print(f"\nüìà Maa≈ü - Performans Korelasyonu (np.corrcoef): {correlation:.4f}")
```

```python
# Matematiksel d√∂n√º≈ü√ºmler
print("\nüßÆ MATEMATƒ∞KSEL D√ñN√ú≈û√úMLER")
print("="*80)

# Normalize etme (0-1 arasƒ±)
maas_normalized = (maas_array - maas_array.min()) / (maas_array.max() - maas_array.min())
print(f"Normalize edilmi≈ü maa≈ü aralƒ±ƒüƒ±: [{maas_normalized.min():.4f}, {maas_normalized.max():.4f}]")

# Standardize etme (Z-score)
maas_standardized = (maas_array - np.mean(maas_array)) / np.std(maas_array)
print(f"Standardize edilmi≈ü maa≈ü - Ort: {np.mean(maas_standardized):.4f}, Std: {np.std(maas_standardized):.4f}")

# Log d√∂n√º≈ü√ºm√º
maas_log = np.log(maas_array)
print(f"Log d√∂n√º≈ü√ºm√º sonrasƒ± ortalama: {np.mean(maas_log):.4f}")
```

## üîü Sonu√ß ve √ñneriler


```python
print("üìã ANALƒ∞Z SONU√áLARI VE √ñNERƒ∞LER")
print("="*80)

# 1. Genel ƒ∞statistikler
print("\n1Ô∏è‚É£ GENEL ƒ∞STATƒ∞STƒ∞KLER:")
print(f"   ‚Ä¢ Toplam √áalƒ±≈üan: {len(df)}")
print(f"   ‚Ä¢ Departman Sayƒ±sƒ±: {df['Departman'].nunique()}")
print(f"   ‚Ä¢ Ortalama Maa≈ü: {df['Maas'].mean():,.2f} TL")
print(f"   ‚Ä¢ Ortalama Performans: {df['Performans_Puani'].mean():.2f}/10")
print(f"   ‚Ä¢ Ortalama Deneyim: {df['Deneyim_Yil'].mean():.1f} yƒ±l")

# 2. Departman Insights
print("\n2Ô∏è‚É£ DEPARTMAN ANALƒ∞Zƒ∞:")
en_yuksek_maas_dept = df.groupby('Departman')['Maas'].mean().idxmax()
en_yuksek_perf_dept = df.groupby('Departman')['Performans_Puani'].mean().idxmax()
en_buyuk_dept = df['Departman'].value_counts().idxmax()

print(f"   ‚Ä¢ En y√ºksek maa≈ü ortalamasƒ±: {en_yuksek_maas_dept}")
print(f"   ‚Ä¢ En y√ºksek performans ortalamasƒ±: {en_yuksek_perf_dept}")
print(f"   ‚Ä¢ En fazla √ßalƒ±≈üana sahip: {en_buyuk_dept}")

# 3. Korelasyon Bulgularƒ±
print("\n3Ô∏è‚É£ ƒ∞Lƒ∞≈ûKƒ∞ ANALƒ∞Zƒ∞:")
maas_deneyim_corr = df['Maas'].corr(df['Deneyim_Yil'])
maas_perf_corr = df['Maas'].corr(df['Performans_Puani'])
print(f"   ‚Ä¢ Maa≈ü-Deneyim korelasyonu: {maas_deneyim_corr:.3f}")
print(f"   ‚Ä¢ Maa≈ü-Performans korelasyonu: {maas_perf_corr:.3f}")

# 4. √ñneriler
print("\n4Ô∏è‚É£ √ñNERƒ∞LER:")
if maas_perf_corr < 0.3:
    print("   ‚ö†Ô∏è Maa≈ü ile performans arasƒ±nda zayƒ±f ili≈üki var.")
    print("      ‚Üí Performansa dayalƒ± √ºcretlendirme sistemini g√∂zden ge√ßirin.")

dusuk_perf = df[df['Performans_Puani'] < 5]
if len(dusuk_perf) > 0:
    print(f"   ‚ö†Ô∏è {len(dusuk_perf)} √ßalƒ±≈üan d√º≈ü√ºk performans g√∂steriyor.")
    print("      ‚Üí Eƒüitim ve geli≈üim programlarƒ± d√º≈ü√ºn√ºlebilir.")

print("\n‚úÖ Analiz tamamlandƒ±!")
```
